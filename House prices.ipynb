```python
for i in train.columns:
    if(train[i].isnull().sum()>0):
        train[i].fillna(method="bfill",inplace=True)
for i in test.columns:
    if(test[i].isnull().sum()>0):
        test[i].fillna(method="bfill",inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
categorical = [f for f in train.columns if train.dtypes[f] == 'object']
train[categorical]=train[categorical].apply(LabelEncoder().fit_transform)
categorical_test = [f for f in test.columns if test.dtypes[f] == 'object']
test[categorical_test]=test[categorical_test].apply(LabelEncoder().fit_transform)
```

```python
from sklearn.preprocessing import MinMaxScaler;
mi=MinMaxScaler();
train_df=pd.DataFrame(mi.fit_transform(train))
train_df.columns=train.columns
train_df.head()
```

```python
test_df=pd.DataFrame(mi.fit_transform(test))
test_df.columns=test.columns
test_df.head()
```

```python
x=train_df.iloc[:,:-1]
y=train_df.iloc[:,-1:]
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
xgb = XGBRegressor()
parameters = {'objective':['reg:squarederror'],
            'learning_rate': [.1],
              'max_depth': [3],
              'min_child_weight': [2],
              'subsample': [0.7],
              'colsample_bytree': [.7],
              'colsample_bylevel':[.7],
              'alpha' : [.05],
              'lambda' : [.3],
              'n_estimators': [2500]}
XGB = GridSearchCV(xgb,
                        parameters,
                        cv = 2)
XGB.fit(x,y)
```

```python
XGB.best_params_
```

```python
XGB.best_score_
```

```python
x_test=test_df
```

```python
y_pred = XGB.predict(x_test)
y_pred
```

```python
submission = pd.DataFrame({
        "Id": test_df.Id,
        "SalePrice": y_pred
    })
submission.to_csv('first_submission.csv', index=False)
```
